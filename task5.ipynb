```python
WITH
    daily AS (
        SELECT DISTINCT
            toDate(`timestamp`) AS day,
            user_id
        FROM dwh.site_activity
    ),
    bounds AS (
        SELECT
            min(day) AS min_day,
            max(day) AS max_day
        FROM daily
    )
SELECT
    addDays(t.dt, -6) AS start_date,
    t.dt AS end_date,
    uniqExact(t.user_id) AS wau_7d
FROM (
    SELECT
        addDays(d.day, offset) AS dt,
        d.user_id
    FROM daily AS d
    ARRAY JOIN range(7) AS offset
) AS t
CROSS JOIN bounds AS b
WHERE
    t.dt >= b.min_day
    AND t.dt <= b.max_day
GROUP BY
    t.dt
ORDER BY
    t.dt;
```
